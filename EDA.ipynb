<h5>Creating a new column of Family Size</h5>

```python


def transform_family_size(num):
    if num == 1:
        return 'alone'
    elif num>1 and num <5:
        return 'small'
    else:
        return 'large'
```

```python
df['Family_size'].apply(transform_family_size)
```

```python
df['Family_type']=df['Family_size'].apply(transform_family_size)
```
